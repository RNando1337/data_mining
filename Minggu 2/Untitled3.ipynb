```python
import numpy as np
import matplotlib.pyplot as mpl
import pandas as pd
```

```python
ds = pd.read_csv("DataWisata.csv")

x = ds.iloc[:, [2,4]].values
y = ds.iloc[:, [0,3]].values
```

```python
print(np.array(x))
```

```python
print(y)
```

```python
from sklearn.compose import ColumnTransformer as CT
from sklearn.preprocessing import OneHotEncoder as OHE
ct = CT(transformers=[('encoder', OHE(sparse=True), [0])], remainder='passthrough') #penamanaan "tes" bebas
x = np.array(ct.fit_transform(x))                            
```

```python
print(x)
```

```python
from sklearn.preprocessing import LabelEncoder as LE
lbl_encode = LE()
y[:, 0] = lbl_encode.fit_transform(y[:, 0])
```

```python
print(y)
```

```python
from sklearn.model_selection import train_test_split as tts
x_train, x_tes, y_train, y_tes = tts(x,y, test_size = 0.2, random_state = 1)
```

```python
print(x_train)
```

```python
print(x_tes)
```

```python
print(y_train)
```

```python
print(y_tes)
```

```python
from sklearn.preprocessing import StandardScaler as SS
sc = SS()
x_train[:, -1:] = sc.fit_transform(x_train[:, -1:])
x_tes[:, -1:] = sc.transform(x_tes[:, -1:])
```
